```python
import numpy as np
import matplotlib.pyplot as plt

#create the figure and two axes (2 rows, 1 column)
fig, (ax1, ax2) = plt.subplots(2,1)

#create a plot of y = sin(x) on the first row

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)
ax1.plot(x1, y1)

#create a plot of y = cos(x)
x2 = np.linspace(0, 4 * np.pi, 100)
y2 = np.cos (x2)
ax2.plot (x2, y2)

#save the figure 
plt.savefig('sin_cos.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
#import another rendering engine
import seaborn as sns

#create the figure and two axes (one row, one column)
fig, ax1 = plt.subplots(1, 1)

#share the x-axis for both the axes
ax2 = ax1.twinx()

#create a plot of y = sin(x) on the first row
x1 = np.linspace(0, 4* np.pi, 100)
y1 = np.sin(x1)
#add a label for the legend and make it blue
function1 = ax1.plot(x1, y1, 'b', label ='Sine')

#create a plot of y = cos(x) on the second row
x2 = np.linspace(0, 4* np.pi, 100)
y2 = np.cos(x2)
#label + make it red
function2 = ax2.plot(x2, y2, 'r', label='Cosine')

#create the legend by first fetching the labels from the function
functions = function1 + function2
labels = [f.get_label() for f in functions]
plt.legend(functions, labels, loc= 3)

#add x and y labels
ax1.set_xlabel('$X$')
ax2.set_ylabel('$Y_2$')
ax1.set_ylabel('$Y_1$')

#ADD A TITLE
plt.title('SINE E COSINE')

#ADJUST THE FIGURE SUCH AS ALL RENDERING COMPONENTS FIT INSIEDE THE FIGURE
plt.tight_layout()

#save the picture
plt.savefig('sin_cos_2')
```
